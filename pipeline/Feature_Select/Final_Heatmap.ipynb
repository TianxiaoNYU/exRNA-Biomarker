```r
library(gplots)
library(pheatmap)
library(scales)
library(reshape2)
library(RColorBrewer)

dm <- as.matrix(read.table("RFE_data.txt", sep = "\t", header = T, row.names = 1))
dm <- t(scale(dm))

pdf("final_heatmap.pdf", 25, 13)
cr <- colorRampPalette(c("blue","white","red"))
heatmap.2(dm,
          scale="row", 
          key=T, keysize=1.1, 
          cexCol=0.9,cexRow=0.8,
          ColSideColors=c(rep("red",99), rep("blue", 50), rep("yellow", 6), rep("green", 36)),
          col=cr(1000),
          density.info="none",trace="none",
          Colv =T,Rowv = T 
)
dev.off()
```
